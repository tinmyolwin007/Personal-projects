# Transform the value, wage and release clause columns into columns of integers

```python
df[['Value','Wage','Release Clause']]
```

```python
def value(x):
    if 'M' in x:
        x = x.replace('M', '')
        x = float(x[1:]) * 1000000
        return round(x)
    elif "K" in x:
        x = x.replace('K', '')
        x = float(x[1:]) * 1000
        return round(x)
    else:
        return int(x[1:])
```

```python
df['Value'] = df['Value'].apply(value)
```

```python
def wage(x):
    if 'K' in x:
        x = x.replace('K', '')
        x = int(x[1:]) * 1000
        return round(x)
    else:
        return int(x[1:])
```

```python
df['Wage'] = df['Wage'].apply(wage)
```

```python
def release_clause(x):
    if 'M' in x:
        x = x.replace('M', '')
        x = float(x[1:]) * 1000000
        return round(x)
    elif "K" in x:
        x = x.replace('K', '')
        x = float(x[1:]) * 1000
        return round(x)
    else:
        return int(x[1:])
```

```python
df['Release Clause'] = df['Release Clause'].apply(release_clause)
```
