# Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
df['Rating_cat'] = df['Rating'].apply(rating)
df.head()
```

# Unique Values From Genre 

```python
df['Genre'].unique()
```

```python
list1=[]
for value in df['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique_list=[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
```

```python
unique_list
```
